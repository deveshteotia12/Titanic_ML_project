# Dealing with NaN

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age',data=test)
```

```python
import cufflinks as cf
```

```python
test[['Age','Pclass']]
```

```python
def myfunc(col):
    Age=col[0]
    Pclass=col[1]
    if (pd.isnull(Age)):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
        
        
```

```python
test['Age']=test[['Age','Pclass']].apply(myfunc,axis=1)
```
